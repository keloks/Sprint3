```python
import matplotlib.image as mpimg
import matplotlib.pyplot as plt

plt.interactive(True)

#%% Functions

def eliminateColour(pic, index):
    
    pic[:,:,index]=0 #Eliminating the layer can cause problems. Resetting all its values to 0 seems more appropiate
    
    return pic

def quadplot(img, imgGB, imgRB, imgRG):
    
    plt.figure()
    
    plt.subplot(221)
    plt.title('Original')
    plt.imshow(img)
    
    plt.subplot(222)
    plt.title('Red channel removed')
    plt.imshow(imgGB)
    
    plt.subplot(223)
    plt.title('Green channel removed')
    plt.imshow(imgRB)
    
    plt.subplot(224)
    plt.title('Blue channel removed')
    plt.imshow(imgRG)
    
    plt.savefig('slicingColours.jpg', bbox_inches='tight')    
```

```python
#%%Exercici 8

img = mpimg.imread('img.jpg') #Also img2 included

imgGB = eliminateColour(img.copy(), 0)

imgRB = eliminateColour(img.copy(), 1)

imgRG = eliminateColour(img.copy(), 2)

quadplot(img, imgGB, imgRB, imgRG)
```
